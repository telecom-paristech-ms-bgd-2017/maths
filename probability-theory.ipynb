# Probability Theory

Let $(A_i)$ to be a countable sequence.

A probability space is a triplet $(\Omega, \Sigma, \mathbb{P})$ where $\Omega$ is a set of events, $\Sigma$ is a $\sigma$-algebra on $\Omega$, and $\mathbb{P}$ a probabiliy measure.

A $\sigma$-algebra on $\Omega$ is a collection of parts of $\Omega$ that includes the empty set, is closed under complement and closed under union (and then closed by intersection):
$$
\begin{array}{ll}
(i) & \forall A \in \Sigma,\; \bar{A} \in \Sigma \\
(ii) & \forall A_i \in \Sigma,\;
\bigcup_{i}A_i \in \Sigma
\end{array}
$$

In particular $\emptyset \in \Sigma$ and $\Omega \in \Sigma$. The set of parts of $\Omega$, $\mathcal{P}(\Omega)$ is a $\sigma$-algebra.  
Often, we take $\Sigma = \mathcal{P}(\Omega)$.

A probabiliy measure $\mathbb{P}$ on $\sigma$-algebra, $\Sigma$, is a measureable application such as:

$$
P(\Omega) = 1
$$

A measureable function $\mu$ on set $E$ application such as:

$$
\begin{array}{ll}
\mathbb{\mu}: & E \rightarrow \mathbb{R^+}\\
& \forall A_i, A_j \subset \Omega, A_i \cap A_j = \emptyset: \\
& \mathbb{\mu}(\coprod_i A_i) = \sum_i \mathbb{\mu}(A_i) 
\end{array}
$$

It is the Ïƒ-additivity property. The $A_i$ are disjoint sets (or mutually exclusive events).

A random variable is a mesureable function:

$$X: \Omega \rightarrow E$$

where $(\Omega, \Sigma, \mathbb{P})$ is probality space and $(E, \mathcal{E})$ is measurable space ($\mathcal{E}$ is a $\sigma$-algebra of $E$). 

The law of probability of $X$ is $\mathbb{P}_X$:

$$\forall B \in \mathcal{E} \quad 
\mathbb{P}_X(B) = \mathbb{P}(X^{-1}(B)) = \mathbb{P}(X \in B)$$

where $X^{-1}(B) = \{ \omega \in \Omega \;|\; X(\omega)\in B \}$, the inverse image of $B$ by $X$.

In particular, if $\Omega$ is finite and:

$$ \forall w \in \Omega, \mathbb{P}(\{\omega\}) = \frac{1}{card(\Omega)}$$

Then:

$$ \forall A \subset \Omega, \mathbb{P}(A) = \frac{card(A)}{card(\Omega)}$$


We have an expression of the mean, the variance and covariance of the indicator function:

$$
\begin{array}{l}
E[\mathbb{1}_A ] = P(A) \\
Var(\mathbb{1}_A) = P(A)(1-P(A)) \\
Cov(\mathbb{1}_A,\mathbb{1}_B) = P(A\cap B) - P(A)P(B)
\end{array}
$$
